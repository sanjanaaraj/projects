```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# web scraping from wikipedia
```

```python
healthcare=pd.read_html("https://en.wikipedia.org/wiki/Healthcare_in_Chennai#")
type(healthcare)
```

```python
hc=healthcare[1]
```

```python
type(hc)
```

```python
#preprocessing data to delete unnecessary columns and convert the number of beds to a suitable format.
```

```python
del hc["Notes"]
```

```python
hc
```

```python
 hc['Number of beds']=hc['Number of beds'].str.replace('[20]','',regex=False)
```

```python
 hc['Number of beds']=hc['Number of beds'].str.replace('[22]','',regex=False)
```

```python
hc['Number of beds']=hc['Number of beds'].str.replace('>','',regex=False)
```

```python
hc.loc[41,'Number of beds']=(135+250)/2
```

```python
hc
```

```python
sns.countplot(x='Hospital type',data=hc,palette='coolwarm')
# the number of govt hospitals continue to predominate
```

```python
hcnew=hc.sort_values(by='Established').head(10)
hcnew
#most of the hospitals established before the 1930's are government setups
```

```python
hc['Number of beds']=hc['Number of beds'].str.replace(',','',regex=False)
hc['Number of beds']=pd.to_numeric(hc['Number of beds'])
              
         
```

```python
mean=hc['Number of beds'].mean()

```

```python
hc[hc['Number of beds']>mean][['Chief specialty','Number of beds','Hospital type']]
#number of beds allocated for government hospitals and specialties such as general medicine and orthopaedics is higher 
```

```python
hc.sort_values(by='Number of beds').head(8)
# taking the hospitals with the least number of beds there is an equal distribution between government and other units
```

```python
sns.barplot(x='Hospital type',y='Number of beds',data=hc,estimator=np.std,palette='magma')
```

```python
hc.pivot_table(values='Number of beds',columns='Established',index='Hospital type')
```

```python
# we can conclude that there has been an increase in the number of private and corporate units aftr the early 1930's , however the number of government hospitals continues to be the maximum.
#taking the number of beds into account , government hospitals continue to provide a greater number and there is a greater allcoation of beds on the higher range for general medicine, obstetrics and orthopaedics . However in the lower range there is an equal distribution in all the sectors and specialities.
```
