```python
# IMPORTING DATASET
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:/Users/yf/Desktop/personality-prediction-from-text-master/personality-prediction-from-text-master/data/training/essays.csv",encoding='cp1252')
data
```

```python
#REMOVING STOPWORDS FOR PREPROCESSING
from nltk.corpus import stopwords
stw=stopwords.words('english')
data['TEXT']=data['TEXT'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stw)]))
data['TEXT']




        
```

```python
#FURTHER PREPROCESSING
import re
from nltk.tokenize import word_tokenize




data['TEXT']=data['TEXT'].apply(lambda x:re.sub(r'''[,-:?!'.1234456]''','',x))
data['TEXT']=data['TEXT'].apply(lambda x:x.lower())
data['TEXT']=data['TEXT'].apply(lambda x:word_tokenize(x))

text=data['TEXT']
text



```

```python
'''IMPORTING GENSIM AND BUILDING VOCAB . WORD2VEC IS A FRAMEWORD WHICH TAKES INTO ACCOUNT THE SEMANTIC WEIGHTAGE OF THE TEXT '''
import gensim
from gensim.models import word2vec
from gensim.models.word2vec import Word2Vec
model=Word2Vec(sg=1,window=6,min_count=2,workers=4)
model.build_vocab(text)

```

```python
vocab=model.wv.index_to_key
vocab
```

```python
model.corpus_count
```

```python
model.train(text,total_examples=model.corpus_count,epochs=model.epochs)
model.save('c:\\Users\\yf\\Desktop.model')
```

```python
#FINDING THE AVERAGE OF WORD EMBEDDINGS TO PRODUCE THE SENTENCE EMBEDDING OF SAME DIMENSIONS
def vector(sent):
    vector_size=model.wv.vector_size
    x=np.zeros(vector_size)
    y=0
    for a in sent:
        if a in vocab:
            y+=1
            x+=model.wv[a]
    x=x/y
    return x
             
```

```python
v=vector('well, right, i, woke, midday, nap, it')
type(v)
```

```python
#APPLYING FUNCTION TO DATA
data['tokens']=data['TEXT'].apply(vector)

data['tokens']
```

```python
#CONVERTING 'Y'&'N' TO 1&0
data['cEXT']=data['cEXT'].map({'y':1,'n':0})
data['cNEU']=data['cNEU'].map({'y':1,'n':0})
data['cAGR']=data['cAGR'].map({'y':1,'n':0})
data['cCON']=data['cCON'].map({'y':1,'n':0})
data['cOPN']=data['cOPN'].map({'y':1,'n':0})
data
```

```python

# DECLARING X AND Y 
x=data['tokens']
y=data[['cEXT','cNEU','cAGR','cCON','cOPN']]
x.dtypes
```

```python
# SPLITTING THE DATA INTO TRAIN AND TEST VARIABLES
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train[0]
```
